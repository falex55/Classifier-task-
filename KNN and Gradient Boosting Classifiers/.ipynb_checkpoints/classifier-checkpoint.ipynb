# Standerdize the variables 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('label',axis=1))
```

```python
 # seperating the features into x and y as independent and dependent features 
```

```python
x = scaler.transform(df.drop('label',axis=1))
```

```python
y=df['label']
```

```python
df_scalar = pd.DataFrame(x,columns=df.columns[:-1])
df_scalar.head()
```

```python
#  From the above pairplot we can easily identify the data is non-linear data so i have used KNN Classifier 
#  also i have implement the data with GradientBoosting Classifier 
```

# Split the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=1)
```

```python
knn_classifier.fit(x_train,y_train)
```

```python
predictions=knn_classifier.predict(x_test)
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
# we can increase the accuracy by claculating the below function to get best neighbors value 
```

```python
error_rate = []
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    predctions_i = knn.predict(x_test)
    error_rate.append(np.mean(predctions_i != y_test))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# from the above figure we are getting less error rate after 10 and after 
# that it is rising again so we will take that value for neighbors
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=11)
```

```python
knn_classifier.fit(x_train,y_train)
```

```python
predictions=knn_classifier.predict(x_test)
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc_classifier=GradientBoostingClassifier(n_estimators=40)
```

```python
gbc_classifier.fit(x_train,y_train)
```

```python
gbc_predictions=gbc_classifier.predict(x_test)
```

```python
print(confusion_matrix(y_test,gbc_predictions))
```

```python
print(classification_report(y_test,gbc_predictions))
```
